```python
import numpy as np
import matplotlib.pyplot as plt

from envs import Maze
from utils_2 import plot_policy, plot_action_values, test_agent
```

```python
env = Maze()
```

```python
action_values = np.zeros((5, 5, 4))
```

```python
def target_policy(state):
    av = action_values[state]
    return np.random.choice(np.flatnonzero(av == av.max()))
```

```python
def exploratory_policy(state):
    return np.random.randint(4)
```

```python
plot_action_values(action_values)
```

```python
plot_policy(action_values, env.render(mode='rgb_array'))
```

```python
#episodes-: number of iterrations
#alpha-: measure the speed at which we update the estimate
#gamma-: discount factor
#initialize the state
#action will be selected by the help of exploratory policy
#execute the action and get the next_state
#next_action will be choosen from target_policy
#q value estimate of both current and next will be saved in action_values
#with both the qsa we will update the current state and action value

def q_learning(action_values, exploratory_policy, target_policy, episodes, alpha=0.1, gamma=0.99):
    
    for episode in range(1, episodes + 1):
        state = env.reset()
        done = False
        
        while not done:
            action = exploratory_policy(state)
            next_state, reward, done, _ = env.step(action)
            next_action = target_policy(next_state)
            
            qsa = action_values[state][action]
            next_qsa = action_values[next_state][next_action]
            action_values[state][action] = qsa + alpha * (reward + gamma * next_qsa - qsa)
    
            state = next_state
```

```python
q_learning(action_values, exploratory_policy, target_policy, 1000)
```
